### Missing Values

```python
# here we will check the percentafe of nan values present in feature
#  1 -step make the list of feature which had missin values
#  writing list comprehension
```

```python
features_with_na = [features for features in dataset.columns if dataset[features].isnull().sum()>1]

# 2- step print the features name and the percentage of missing values

for features in features_with_na:
    print(features, np.round(dataset[features].isnull().mean(), 4), '% missing values')
```

### Since there are lot of missing values, we need to find the relationship between missing values and Salse Price

1. lets plot some diagrame for this relationship

```python
for feature in features_with_na:
    data = dataset.copy()
    
#     lets make a variable if which indicate 1 if the obeservation is missing otherwise 0
    data[feature] = np.where(data[feature].isnull(), 1, 0)
#     lets calculate the mean salseprice where the information is present or missing
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feature)
    plt.show()
```

Here with relationship between missing value and dependent variable is clearly visible. so we need to replace these nan values with something meaningful which we will do in the fearutre engg. part

### Numerical Features

creating a variable to get numerical features

```python
# list of numerical features
numerical_features = [feature for feature in dataset.columns if dataset[feature].dtypes != 'O']

print("Numbers of Numerical Features", len(numerical_features))

# visualize the numerical features
dataset[numerical_features].head()
```

### Temporal Variable (Like Date Time)


there are 4 year variables.

From the Dataset we have 4 year variables. We have extract information from the datetime variables like no of years or no of days. One example in this specific scenario can be difference in years between the year the house was built and the year the house was sold. We will be performing this analysis in the Feature Engineering which is the next part.

```python
year_feature = [feature for feature in numerical_features if 'Yr' in feature or 'Year' in feature]

year_feature
```

```python
# lets explore the content of these features

for feature in year_feature:
    print(feature, dataset[feature].unique())
```

```python
# lets analyze the temporal variabels 
# we will check wether there is relation between year the house is sold and sales price

dataset.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('Year Sold')
plt.ylabel('Sale Price')
plt.title('Year Sold vs Sale price')
```

```python
#  as we can see as the year increasing the sale price is decreasing 
```

```python
year_feature
```

```python
# here we will compare the difference between all years feature with sales price

for feature in year_feature:
    if feature != 'YrSold':
        data = dataset.copy()
        ## we will capture the difference between year variable and year the house was sold for
        data[feature] = data['YrSold']- data[feature]
        
        plt.scatter(data[feature], data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.show()
    
```

### Exploratory Data Analysis part 2

```python
for feature in conti_feature:
    data=dataset.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        data['SalePrice']=np.log(data['SalePrice'])
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalesPrice')
        plt.title(feature)
        plt.show()
```

### outliers

```python
for feature in conti_feature:
    data = dataset.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature] = np.log(data[feature])
        data.boxplot(column = feature)
        plt.ylabel(feature)
        plt.title(feature)
        plt.show()
```
